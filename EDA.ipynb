From the above operation, we confirm that they were no missing values in the data. We can also visualise the missing values from the below plot.

**Correlation** is a statistic that measures the degree to which two variables move in relation to each other. The correlation between the variables can be calculated using the below python function. If there exists a positive value output between the two variables, then they are **positively** **correlated** & the vice- versa is **negatively** **correlated**.

```python
data.corr()
```

```python
# Let's see the correlation matrix 
plt.figure(figsize = (20,10))        # Size of the figure
sns.heatmap(data.corr(),annot = True)
```

From the above table, we observe that **win** **points**, **kill** **points** are highly positively correlated with a correlation value of **0.9834**. So, we can drop one column either win points or kill points. Here, I drop the **kill** **points** column. Also, **rank** **points** is highly negatively correlated with a value of **-0.99**.

**Killplace** and **win place** are correlated positively by **0.72**

**Univariate** **Plots:**

Univariate data visualization plots help us comprehend the enumerative properties as well as a descriptive summary of the particular data variable. These plots help in understanding the location/position of observations in the data variable, its distribution, and dispersion. Uni-variate plots are of two types: 
1. Enumerative plots and
2. Summary plots

**Scatter** **Plot**

```python
plt.scatter(data.index,data['kills'])
```

```python
plt.scatter(data.index,data['headshotKills'])
```

```python
plt.scatter(data.index,data['heals'])
```

**Strip** **Plot** - The strip plot is similar to a scatter plot. It is often used along with other kinds of plots for better analysis. It is used to visualize the distribution of data points of the variable.

```python
sns.stripplot(y=data['heals'])
```

```python
sns.stripplot(y=data['roadKills'])
```

**Histogram** - A histogram is a graphical display of data using bars of different heights. In a histogram, each bar groups numbers into ranges. Taller bars show that more data falls in that range. A histogram displays the shape and spread of continuous sample data.

```python
plt.hist(data['boosts'],color='grey')
```

This shows how much data is categorised into different blocks.
